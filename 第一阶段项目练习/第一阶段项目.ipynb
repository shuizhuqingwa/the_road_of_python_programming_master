### 要求
- 打印三级菜单
- 可返回上一级
- 可随时退出程序

```python
dict1={'书籍':{'电子书':['python基础','java基础'],'纸质书':['疯狂python','疯狂java']},
          '电器':{'大电器':['空调','冰箱'],'小电器':['风扇','灯泡']},
          '家具':{'大家具':['床','沙发'],'小家具':['桌子','椅子']}
         }
   
def print_dir(dict2):
    for k in dict2.keys():
        print(k)     
temp_dir_list=[]   
tag = True
while tag:   
    print_dir(dict1)
    if dict1  not in temp_dir_list :
        temp_dir_list.append(dict1)
    choice = input("选择当前目录，【exit】 退出，【back】返回上一层 >>:").strip() #选择当前目录
    if choice == 'exit':
        tag=False
        break
    if choice == 'back':
        if len(temp_dir_list) == 1:
            print("当前已是第一层,请选择目录往下打印")
            continue
        else:
            temp_dir_list.remove(temp_dir_list[-1])
            dict1=temp_dir_list[-1]
            print('返回上一层,当前目录>>:')
    if choice != 'exit' and  choice != 'back':
        if choice not in dict1.keys():
            print("输入项非法,请重新输入>>:")
    if choice in dict1.keys():
        if not isinstance(dict1[choice],dict):
            print('已到最底层，当前目录内容：',dict1[choice])
            sub_tag = True
            while sub_tag:
                choice2=input('输入【back】返回上一层,【exit】 退出>>:').strip()
                if choice2 != 'back' and choice2 !='exit':
                    print("你输入的值非法，请输入【back】返回上一层 or 【exit】 退出")     
                else:
                    if choice2 ==  'exit':
                        tag = False
                        sub_tag=False
                        break
                    else:
                        print("返回上一层,当前目录>>:")
                        sub_tag = False
                        dict1=temp_dir_list[-1]
            continue
        dict1 = dict1[choice]
        
```

### 简单版商城

```python

# 检查用户注册信息
def user_info_regist_check(file_path,user_regist_info):
    if inquire_user_info(file_path,user_regist_info[0]):
        print("该用户已存在")
        return None
    if user_regist_info[1] !=user_regist_info[2]:
        print("两次输入的密码不一致，请重新输入")
        return None
    return 1
        
#查询用户信息并返回  
def inquire_user_info(file_path,user_name):
    data = reload_file(file_path)
    temp_dir ={}
    temp_list=[]
    for line in data:
        temp_list =line.split('|')
        temp_dir[temp_list[0]]=temp_list[1:]
    if user_name not in temp_dir.keys():
        return None
    else:
        temp_list=temp_dir[user_name]
        temp_list.insert(0,user_name)
        return temp_list

#用户注册
def user_regist(file_path,user_regist_info):
    data_list = reload_file(file_path)
    str1 =''
    for info in user_regist_info:
        info = str(info)+'|'
        str1 +=info
    str1=str1.strip('|')
    data_list.append(str1)
    with open(file_path,'w+') as f:
        for line in data_list:
            f.write(line+'\n')
    return 1
#用户信息更新   
def user_info_update(file_path,user_info):
    data=[]
    data_list = reload_file(file_path)
    for line in data_list:
        temp_line = line
        temp_list =line.split('|')
        if temp_list[0]==user_info[0]:
            temp_list[-1]=user_info[-1]
            temp_line=''
            for info in temp_list:
                info = str(info)+'|'
                temp_line +=info
            temp_line=temp_line.rstrip('|')
        data.append(temp_line)
    write_file(file_path,data)

#查询商城物品
def inquire＿shop(shop,goods):
    num =shop[goods][0]
    price = shop[goods][1]
    return num,price
#重新加载文件
def reload_file(file_path):
    data =[]
    with open(file_path,'r+') as f:
        for line in f.readlines():
            data.append(line.strip())
    return data
#写文件
def write_file(file_path,data):
    with open(file_path,'w+') as f:
        for line in data:
            f.write(line+'\n')
#加入黑名单
def add_black_list(file_path,username):
    data = reload_file(file_path)
    data.append(username)
    write_file(file_path,data)
    
#查询黑名单信息
def inquire_black_user_info(file_path,user_name):
    black_list = reload_file(file_path)
    temp_dir ={}
    for line in  black_list:
        temp_dir[line]= 1
    if user_name in temp_dir.keys():
        return 1
    else:
        return None
    
exit_flag = False

#简单版本商城
shop={'iphone':[100,6000],'macbook':[100,10000],'bike':[100,600]}
#物品映射编码
shop_map ={'1':'iphone','2':'macbook','3':'bike'}
buy_list=[]
shopping = True

file_path ='.\用户信息.txt'
black_file_path ='.\\blacklist.txt'


print('='*20+ '欢迎进入商城'+ '='*20)
user_name = input("[请输入用户名》》：]").rstrip()
if inquire_black_user_info(black_file_path,user_name):
    print("该用户已被锁，禁止登陆！")
else:
    if not inquire_user_info(file_path,user_name):
        choice=str(input("该用户未注册，请先注册，输入[1] 注册，输入[Exit] 退出 >>:"))
        if choice == '1':
            user_name =input("请输入用户名>>:").rstrip()
            user_password1 =input("首次输入密码>>:").rstrip()
            user_password2 =input("再次输入密码>>:").rstrip()
            user_regist_info=[user_name,user_password1,user_password2]
            while True:
                if not user_info_regist_check(file_path,user_regist_info):
                    choice2=input("'继续注册 'Y' OR 退出 'N' >>:").rstrip()
                    if choice2 =='Y':

                        user_name =("'请输入用户名>>:").rstrip()
                        user_password1 =("'首次输入密码>>:").rstrip()
                        user_password2 =("'再次输入密码>>:").rstrip()
                        user_regist_info=[user_name,user_password1,user_password2]
                    else:
                        exit_flag = True
                        break
                else:
                    user_info =[user_name,user_password1,0]
                    if user_regist(file_path,user_info):
                        print("恭喜你，注册成功")
                        while True:
                            acct_balance = input("请存入账户金额 》》：")
                            if not str(acct_balance).isdigit():
                                print("输入的金额必须为数字")
                            else:
                                user_info[-1] = acct_balance
                                user_info_update(file_path,user_info)
                                break
                    else:
                        print('注册失败，请联系客服')
                    break
        else:
            exit_flag = True
            pass
    else:
        user_info_temp = inquire_user_info(file_path,user_name)
        print('请登录')
        user_password=input('请输入密码>>:').strip()
        err_count =0
        while True:
            if user_password !=user_info_temp[1]:
                err_count +=1
                if err_count >= 3:
                    print("密码错误次数超过3次，用户已被锁无法登陆!!!")
                    add_black_list(black_file_path,user_name)
                    exit_flag = True
                    break
                print('密码错误，错误次数超过3次用户将被锁禁止登陆，您还可以尝试输入密码 {} 次'.format(3-err_count))
                user_password=input('请重新输入密码>>:')    
            else:
                print('登录成功')
                break
    if exit_flag:
        print('退出')
    else:
        print('='*50)
        print("开始购物")
        user_info = inquire_user_info(file_path,user_name)
        buy_list =[]
        buy_dir ={}
        while shopping:
            print('1 - iphone\n2 - macbook\n3 - bike\nE - 退出')
            buy_choice = input("选择对应商品编号").strip()
            if buy_choice == 'E':
                print('结束购物')
                shopping = False
                break
            if shop_map.get(buy_choice):
                buy_sub_list=[]
                goods_name = shop_map.get(buy_choice)
                goods_num = shop.get(goods_name)[0]
                goods_price = shop.get(goods_name)[1]
                print('{}该物品还有{}件：'.format(goods_name,goods_num))
                while shopping:
                    buy_num = input("请输入购买数量，输入 E 退出").strip()
                    if buy_num == 'E':
                        shopping = False
                        break
                    if str(buy_num).isdigit():
                        if int(buy_num) > goods_num:
                            print("你要购买的物品数量不足")
                        elif int(buy_num) == 0:
                            print("购买物品数量不可为0")
                        else:
                            total_price = int(buy_num) * goods_price
                            if int(user_info[2]) < total_price:
                                print("账户可用余额不足，余额：{} RMB".format(user_info[2]))
                                other_choice = input("充值 Y 或 N退出》》：")
                                if other_choice =='Y':
                                    while True:
                                        acct_balance = input("请存入账户金额 》》：")
                                        if not str(acct_balance).isdigit():
                                            print("输入的金额必须为数字")
                                        else:
                                            user_info[-1] = acct_balance
                                            user_info_update(file_path,user_info)
                                            break
                                else:
                                    shopping = False
                            else:
                                user_acct_bal = int(user_info[2])
                                user_acct_bal -= total_price
                                user_info[2] = user_acct_bal
                                user_info_update(file_path,user_info)
                                goods_num -= int(buy_num)
                                buy_sub_list.append(goods_name)
                                buy_sub_list.append(buy_num)
                                buy_sub_list.append(total_price)
                                buy_list.append(buy_sub_list)
                                print("购买了商品{} 数量{}".format(goods_name,buy_num))
                                break
                                
                    else:
                        print("你输入的值不为数字,请重新输入")
                        
            else:
                print('没有这个物品，请重新选择')
            buy_again_choice=input("是否继续购买，Y/N 》》：")
            if buy_again_choice.upper() != 'Y':
                shopping = False
                print("结束购物")
                if len(buy_list) !=0:
                    print("您购买了以下商品")
                    for i in buy_list:
                        print("{} {}件，消费：{} RMB".format(i[0],i[1],i[2]))  
                    print("账户余额：RMB",user_info[2]) 
                else:
                    print("您本次没有购买任何商品")
                
                

                
                
            
        

        
    
                    
                    
        


    
```
